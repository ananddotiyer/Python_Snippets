The following code converts any image file to an excel file, with the pixel intensities exported to it.

```python
import numpy as np
import pandas as pd
import cv2

pic = cv2.imread(r'files\Bambu1.png')

pic_arr = np.asarray(pic)
```

```python
pic_mean_arr = np.round(pic_arr.mean(axis=2))
print(pic_mean_arr.shape)
```

```python
## convert your array into a dataframe
df = pd.DataFrame (pic_mean_arr)
```

```python
## save to xlsx file
filepath = r'files\Bambu1.xlsx'

df.to_excel(filepath, index=False, header=False)
```
